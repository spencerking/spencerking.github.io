```python
import hail as hl
```

# Prepare variant-disease associations

We need to add a locus column to the tsv file so that we can use it to annotate our variant data.

all_variant_disease_associations.tsv is downloaded from DisGeNET.

```python
ht = hl.import_table('all_variant_disease_associations.tsv', impute=True)
```

```python
ht.describe()
```

```python
import pandas as pd
```

```python
df = pd.read_csv('all_variant_disease_associations.tsv', sep='\t')
```

```python
df
```

```python
df['locus'] = df['chromosome'].astype(str) + ':' + df['position'].astype(str)
```

```python
df
```

```python
df.to_csv('variants.tsv', sep='\t', index=False)
```

# Annotate variants with disease names

AJtrio_HiSeq300X_cortex_variants_GRCh37_09042015.vcf is downloaded the Genome in a Bottle consortium's FTP server.

```python
mt = hl.import_vcf('AJtrio_HiSeq300X_cortex_variants_GRCh37_09042015.vcf', reference_genome='GRCh37')
```

```python
ht = hl.import_table('variants.tsv', impute=True).key_by('locus')
```

```python
mt.describe()
```

```python
mt = mt.annotate_rows(locus_str = hl.str(mt.locus.contig) + hl.str(':') + hl.str(mt.locus.position))
```

```python
mt.locus_str.show()
```

```python
mt = mt.annotate_rows(diseaseName = ht[mt.locus_str].diseaseName)
```

```python
mt.describe()
```

```python
mt.count()
```

```python
mt = mt.filter_rows(hl.is_defined(mt.diseaseName)) 
```

```python
mt.count()
```

```python
mt.diseaseName.show()
```
